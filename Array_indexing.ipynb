üìå #Slicing allows selecting a range of elements from a list or array using:
    [start : end : step]

‚úÖ **Rules**:
- start:   Starting index (default = 0)
- end:     Ending index (not included, default = len(array))
- step:    Step size (default = 1)

üß† Tip:
- Negative step = reverse order

```python
arr[1:5]
```

```python
arr[:5]
```

```python
arr[1:]
```

```python
arr[1:5:2]
```

**Boolean Indexing**
üß† Boolean Indexing in NumPy
‚úÖ What is Boolean Indexing?
Boolean indexing lets you filter elements in an array using True/False conditions.

Instead of giving index numbers, you give a condition, and NumPy returns all elements that match!

üìå Key Syntax:
array[condition]


```python
arr=np.arange(11,21)
arr
```

```python
bool_index=arr%2 ==0
bool_index
```

```python
even_ar = arr[bool_index]
even_ar
```
